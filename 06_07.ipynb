1- Número Invertido
Leia um número inteiro positivo e mostre seus dígitos na ordem inversa.
Exemplo:
Digite um número: 12345
54321

2- Contador de Dígitos
Peça um número inteiro positivo.
Informe quantos dígitos ele possui.
Exemplo:
Digite um número: 987654
Quantidade de dígitos: 6

3- Cofrinho
Uma pessoa deseja guardar R$ 10.000.
A cada mês ela deposita um valor informado pelo usuário.
O programa deve informar:
.
Quantos meses foram necessários.
O total acumulado.
Em qual mês a meta foi atingida.

4- Caixa Registradora
O programa recebe:
preço do produto
quantidade comprada
Ao final mostra:
subtotal
desconto (0%, 5%, 10% ou 15%)
valor final
O programa continua registrando produtos até que seja digitado preço igual a zero.

```python
#1

numero = int(input("Digite um número inteiro: "))

numero_reverso = 0

numero_aux = abs(numero)

while numero_aux > 0:

   digito = numero_aux % 10

   numero_reverso = numero_reverso * 10 + digito

   numero_aux = numero_aux // 10

if numero < 0:

   numero_reverso = -numero_reverso

print("O número com os dígitos em ordem inversa é:", numero_reverso)
```

```python
#2
numero = input("Digite um número: ")

if int(numero) > 0:
    quantidade = len(numero)
    print(f"Quantidade de dígitos: {quantidade}")
else:
    print("Por favor, digite um número inteiro positivo.")
```

```python
#3

meta = 10000.0
total_acumulado = 0.0
meses = 0

while total_acumulado < meta:
    meses += 1

    deposito = float(input(f"Informe o valor do depósito para o mês {meses}: R$ "))

    while deposito < 0:
        print("Por favor, insira um valor válido (maior ou igual a zero).")
        deposito = float(input(f"Informe o valor do depósito para o mês {meses}: R$ "))

    total_acumulado += deposito


print("\n--- Meta Atingida! ---")
print(f"Quantos meses foram necessários: {meses}")
print(f"O total acumulado: R$ {total_acumulado:.2f}")
print(f"Em qual mês a meta foi atingida: Mês {meses}")
```

```python
#4


subtotal = 0.0

print("--- CAIXA REGISTRADORA ---")

while True:
    preco = float(input("Digite o preço do produto (ou 0 para encerrar): "))

    if preco == 0:
        break

    quantidade = int(input("Digite a quantidade comprada: "))

    subtotal += preco * quantidade

if subtotal >= 1000.00:
    desconto_percentual = 15
elif subtotal >= 500.00:
    desconto_percentual = 10
elif subtotal >= 200.00:
    desconto_percentual = 5
else:
    desconto_percentual = 0

valor_desconto = subtotal * (desconto_percentual / 100)
valor_final = subtotal - valor_desconto

print("\n--- RESUMO DA COMPRA ---")
print(f"Subtotal: R$ {subtotal:.2f}")
print(f"Desconto aplicado: {desconto_percentual}% (R$ {valor_desconto:.2f})")
print(f"Valor Final: R$ {valor_final:.2f}")
```
